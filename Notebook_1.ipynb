```
#Pattern 5

x=1
c=1
for i in range(0,4):
  x=c+i
  for j in range (0,i+1):
    print(x, end = ' ')
    x-=1
    c+=1
  print()
```

```
#Pattern 6

from math import factorial
n = 7
for i in range(n):
	for j in range(i+1):
		print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")
	print()
```

```
#Pattern 7

rows = 5
for i in range(1, rows + 1):  
    for j in range(1, rows + 1):  
        if j <= i:  
            print(i, end=' ')  
        else:  
            print(j, end=' ')  
    print()  
```

```
#Pattern 9

rows = 6 
k = 2 * rows - 2    
for i in range(rows, -1, -1):    
    for j in range(k, 0, -1):  
        print(end=' ')  
    k = k + 1  
    for j in range(0, i + 1):  
        print('*', end = ' ')  
    print()  
```

```
#Pattern 10

n = 10 
m = (2 * n) - 2  
for i in range(0, n):  
    for j in range(0, m):  
        print(end=" ")  
    m = m - 1
    for j in range(0, i + 1):   
        print("* ", end=' ')  
    print(" ")  
```

```
#Pattern 13

rows = 5
k = 2 * (rows - 1)
for i in range(1,rows+1):
  for j in range(k,0,-1):
    print(end = ' ')
  k=k-2
  for j in range(i+1, 1 , -1):
    print('* ', end = '')
  print()

rows = rows - 1
k = (rows-2)
for i in range(rows,0,-1):
  for j in range(0,k):
    print(end = ' ')
  k=k+2
  for j in range(1,i+1):
    print('* ', end = '')
  print()
```

```
#Pattern 14

row = 5
# Upper-half
for i in range(row, 0, -1):
    for j in range(row-i):
        print(" ", end="")
    for j in range(1, 2*i):
        print("*", end="")
    print()

# Lower-half
for i in range(2, row+1):
    for j in range(row-i):
        print(" ", end="")
    for j in range(1, 2*i):
        print("*", end="")
    print()
```

```
#Pattern 15

for i in range(1,9):
    for j in range(1,17):
        if(j<=9-i or j>=8+i):
            print("*", end="")
        else:
            print("_",end="")
    print()
```
